```python
from bs4 import BeautifulSoup
import requests

page = requests.get("https://en.wikipedia.org/wiki/Python_(programming_language)")
soup = BeautifulSoup(page.content, 'html.parser')

print(soup.prettify())
```

```python
list(
    soup\
    .find("div", attrs={"class": "reflist columns references-column-width"})\
    .children
)
```

```python
test = list(soup.find("div", attrs={"class": "reflist columns references-column-width"}).descendants)\
[4]
test.name
```

```python
data = soup\
    .find('div', attrs={"class":"reflist columns references-column-width"})\
    .descendants

for d in data:
    if d.name=="span" and d.get("class","")==["reference-text"]:
        print(d.text,"\n")
```

```python
data = soup.find("div", attrs={"class":"reflist columns references-column-width"})
for a in data.find_all("a"):
    if a["href"].startswith("http"):
        print(a["href"])
```

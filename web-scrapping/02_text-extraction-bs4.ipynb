```python
from bs4 import BeautifulSoup
import requests
import re

page = requests.get("https://en.wikipedia.org/wiki/Python_(programming_language)")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
elements = soup.find("body").find_all("p")
elements
```

```python
def clean_html(html):
    exp = re.compile("<.*?>")
    text = re.sub(exp,"",html)
    return text
```

```python
for e in elements:
    print(clean_html(str(e)))
```

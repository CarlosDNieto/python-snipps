```python
from bs4 import BeautifulSoup
import requests
import re

page = requests.get("https://en.wikipedia.org/wiki/Python_(programming_language)")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# all the external links (not wikipedia links)
for link in soup.find_all("a", attrs={"href": re.compile("^http://")}):
    print(link.get("href"))
```

```python
# only wikipedia links
baselink = "https://en.wikipedia.org"

for link in soup.find_all("a"):
    if str(link.get("href"))[0] == "/" and str(link.get("href"))[1] != "/":
        print(baselink + link.get("href"))
```

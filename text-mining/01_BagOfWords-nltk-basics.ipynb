```python
import nltk
import string


from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords

text = open("/Users/carlosnieto/Desktop/text-mining/text.txt", "r").read()
```

```python
# download this to not get an error
nltk.download('punkt')
nltk.download('stopwords')
```

```python
sentences = sent_tokenize(text)
print(sentences)
```

```python
words = word_tokenize(text)
print(words)
```

```python
fdist = FreqDist(words)
fdist.most_common(20)
```

```python
stop_words = stopwords.words("spanish")
stop_words
```

```python
words2 = [x for x in words if x not in stop_words]
words2
```

```python
words = word_tokenize(text.lower())
stop_words = stopwords.words("spanish")
words2 = [x for x in words if x not in stop_words + list(string.punctuation)]
fdist = FreqDist(words2)
fdist.most_common(20)
```

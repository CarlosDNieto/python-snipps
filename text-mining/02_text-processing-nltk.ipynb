```python
import string
from nltk.stem import SnowballStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.probability import FreqDist

text = open("/Users/carlosnieto/Desktop/text-mining/text.txt", "r").read()
stop_words = stopwords.words("spanish")

words = [x for x in word_tokenize(text.lower()) if x not in stop_words+list(string.punctuation)]
```

### ¿What is an ``Stemmer``?

```python
stemmer = SnowballStemmer("spanish")
```

```python
roots = []
for word in words:
    roots.append(stemmer.stem(word))
```

```python
roots
```

```python
FreqDist(roots).most_common(20)
```

```python
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)
```
